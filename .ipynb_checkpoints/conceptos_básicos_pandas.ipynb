Cambiar el type de una columna product_price. Pero tiene un simbolo de $ que debo eliminar primero. La función srt.replace siempre me falla por alguna razon. Mejor usar la función map

Los valores tienen una coma que no es reconocido como separador de miles, primero cambiar la coma por punto y luego convertirlo a float

```python
# Eliminar un caracter
df['product_price'] = df['product_price'].str.replace('$', '').str.replace(',', '').str.replace('.', '')
```

```python
# Verifico los valores de la columna product_price y además veo la frecuencia de cada uno
df.product_price.value_counts()
```

```python
df['product_price'] = df['product_price'].astype(float)
```

```python
# ver el tipo de las columnas 
print(df.dtypes)
```

Hacer lo mismo con la columnas: purchase_price

```python
df['purchase_price'] = df['purchase_price'].str.replace('$', '').str.replace(',', '').str.replace('.', '')
df.purchase_price.value_counts()
```

```python
df['purchase_price'] = df['purchase_price'].astype(float)
```

```python
df.purchase_price.value_counts()
```

```python
# Aplicar función a una columna
def unidad(op1,op2):
    return op1 / op2

df['unit_price'] = df.apply(lambda row: unidad(row['purchase_price'], row['purchase_size']), axis=1)
```

Este código creará una nueva columna unit_price que contiene los resultados de aplicar la función unidad a cada fila de las columnas product_price y purchase_price.

La función lambda se utiliza para aplicar la función unidad a cada fila del DataFrame, pasando como argumentos los valores de product_price y purchase_price. El argumento axis=1 indica que la función se aplica a lo largo de las filas del DataFrame.
