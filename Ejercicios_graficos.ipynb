## Ejercicio para comprender gráficos

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = {'Year': [2010, 2011, 2012, 2013, 2014, 2015],
        'Sales': [10, 20, 30, 40, 50, 60],
        'Expenses': [8, 15, 25, 30, 40, 45]}
df = pd.DataFrame(data)
```

```python
df.head()
```

1. Gráficos de líneas
Los gráficos de líneas son útiles cuando se desea mostrar la tendencia de una serie de datos a lo largo del tiempo o en una secuencia ordenada. Para crear un gráfico de líneas en Matplotlib, podemos usar la función plot():

```python
plt.plot(df['Year'], df['Sales'])
plt.show()
```

2. Gráficos de barras
Los gráficos de barras son útiles cuando se desea comparar diferentes categorías o grupos de datos. Para crear un gráfico de barras en Matplotlib, podemos usar la función bar():

```python
plt.bar(df['Year'], df['Sales'])
plt.show()
```

3. Gráficos de dispersión
Los gráficos de dispersión son útiles cuando se desea mostrar la relación entre dos variables continuas. Para crear un gráfico de dispersión en Matplotlib, podemos usar la función scatter():

```python
plt.scatter(df['Expenses'], df['Sales'])
plt.show()
```

4. Gráficos de torta
Los gráficos de torta son útiles cuando se desea mostrar la proporción de cada categoría en un conjunto de datos. Para crear un gráfico de torta en Matplotlib, podemos usar la función pie():

```python
plt.pie(df['Sales'], labels=df['Year'])
plt.show()
```

Por ejemplo, si queremos comparar la cantidad de descargas de cada software en diferentes países, podríamos utilizar un gráfico de barras. En este caso, cada barra representa la cantidad de descargas de un software en un país determinado. Para crear este gráfico, podríamos utilizar la función bar de Matplotlib:

```python
paises = ['Argentina', 'Brasil', 'Chile', 'Colombia', 'México']
descargas_basic = [100, 200, 150, 180, 120]
descargas_premium = [50, 100, 70, 90, 60]
descargas_pro = [20, 40, 30, 35, 25]
```

```python
plt.bar(paises, descargas_basic, label='Basic')
plt.bar(paises, descargas_premium, bottom=descargas_basic, label='Premium')
plt.bar(paises, descargas_pro, bottom=[sum(x) for x in zip(descargas_basic, descargas_premium)], label='Pro')
# Añadir leyenda y títulos
plt.legend()
plt.title('Descargas por software y país')
plt.xlabel('País')
plt.ylabel('Cantidad de descargas')

# Mostrar gráfico
plt.show()
```
